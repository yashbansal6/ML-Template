```python
import pandas as pd
from sklearn import preprocessing
```

```python
df = pd.read_csv("../input/cat-in-the-dat-ii/train.csv")
```

```python
df.head()
```

```python
df.bin_4.value_counts()
```

```python
lbl = preprocessing.LabelEncoder()
lbl.fit(df.bin_0.values)
```

```python
#df.bin_0.values.tolist() : to see if we have NaN values
df.bin_0.values
```

```python
#error due to Nan
lbl.transform(df.bin_0.values)
```

Correct way is to fillna

```python
lbl.fit(df.bin_0.fillna(-1).values)
```

```python
df.bin_0.fillna(-1).values
```

```python
lbl.transform(df.bin_0.fillna(-1).values)
```

```python
lbl = preprocessing.LabelBinarizer()
lbl.fit(df.bin_0.fillna(-1).values)
```

```python
lbl.transform(df.bin_0.fillna(-1).values)
```

```python
df.bin_0
```

```python
lbl = preprocessing.OneHotEncoder()
#expects 2D array and we pass 1D as only 1 column is passed-> reshape
lbl.fit(df.bin_0.fillna(-1).values.reshape(-1,1))
```

```python
lbl.transform(df.bin_0.fillna(-1).values.reshape(-1,1))
```

```python
lbl = preprocessing.OneHotEncoder()
#no need to reshape as 2 columns passed
lbl.fit(df[["bin_0","bin_1"]].fillna(-1))
```

```python
lbl.transform(df[["bin_0","bin_1"]].fillna(-1))
```
